```python
def get_stock_data(stock_list):
    # Define "good" values for each metric
    good_eps = 1.0  # Positive EPS
    good_pe = 20.0  # Reasonable P/E ratio
    good_peg = 1.0  # PEG ratio close to 1
    good_pb = 3.0  # Reasonable price to book value
    good_dpr = 50.0  # Dividend payout ratio less than 50%
    good_dy = 2.0  # Dividend yield of 2%

    # Loop through each stock and get the financial metrics
    for stock in stock_list:
        try:
            ticker = yf.Ticker(stock)
            info = ticker.info

            # Check if all data is available
            if all(key in info for key in ['trailingEps', 'trailingPE', 'pegRatio', 'priceToBook', 'payoutRatio', 'dividendYield']):
                eps = info['trailingEps']
                pe = info['trailingPE']
                peg = info['pegRatio']
                pb = info['priceToBook']
                dpr = info['payoutRatio'] * 100  # Convert to percentage
                dy = info['dividendYield'] * 100  # Convert to percentage

                # Screen the stocks
                if eps > good_eps and pe < good_pe and abs(peg - good_peg) < 0.5 and pb < good_pb and dpr < good_dpr and dy > good_dy:

                    return (tw.fill(
                        f"{stock} passed the screening with EPS of {eps}, P/E of {pe}, PEG of {peg}, P/B of {pb}, DPR of {dpr}%, and DY of {dy}%."))
        except Exception as e:
            print(f"Could not download data for {stock}.")
            print(e)
```
